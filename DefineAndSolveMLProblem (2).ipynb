At first, we use LINEAR REGRESSION.

```python
# YOUR CODE HERE
# data preparation!

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=123)

# model we use: Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# set predictions
lr_y_preds = model.predict(X_test)
#get the rmse
lr_rmse = mean_squared_error(y_test, lr_y_preds)
#r2 score
lr_r2 = r2_score(y_test, lr_y_preds)

## PRINT OUT THE PREDS
print('[LR] Root Mean Squared Error: {0}'.format(lr_rmse))
print('[LR] R2: {0}'.format(lr_r2))
```

Then, we compare to DECISION TREES OPTIMIZED THROUGH GRID SEARCH.

```python
# code for decision trees -- first try to get the best parameters
param_grid = {"max_depth": [4,8], "min_samples_leaf": [25,50]}
dt_regressor = DecisionTreeRegressor()
dt_grid = GridSearchCV(dt_regressor,param_grid, cv =3, scoring = 'neg_root_mean_squared_error')
dt_grid_search = dt_grid.fit(X_train, y_train)
rmse_DT = -1 * dt_grid_search.best_score_
print("[DT] RMSE for the best model is : {:.2f}".format(rmse_DT) )
dt_best_params = [dt_grid_search.best_estimator_.max_depth, dt_grid_search.best_estimator_.min_samples_leaf]# YOUR CODE HERE
print("Best Params")
print(dt_best_params)
```

```python
# run, create, and fit the DTRegressor Model
dt_model = DecisionTreeRegressor(max_depth = dt_best_params[0], min_samples_leaf = dt_best_params[1])
dt_model.fit(X_train, y_train)
```

```python
# MAKE PREDICTIONS ON DTRegressor
y_dt_pred = dt_model.predict(X_test)
dt_rmse = mean_squared_error(y_test, y_dt_pred)
dt_r2 = r2_score(y_test, y_dt_pred)
print('[DT] Root Mean Squared Error: {0}'.format(dt_rmse))
print('[DT] R2: {0}'.format(dt_r2))
```

From this, we can see that LINEAR REGRESSION PERFORMS BETTER THAN DECISION TREE REGRESSION EVEN ON THE BEST MODEL.


But what happens if we stack these two models? 

```python
# set the estimators and set up
estimators = [("LR", LinearRegression()),("DT", DecisionTreeRegressor(max_depth=8, min_samples_leaf=50)) 
             ]
stacking_model = StackingRegressor(estimators = estimators, passthrough = False)
stacking_model.fit(X_train, y_train)
```

```python
# make predictions
stacking_pred = stacking_model.predict(X_test)
stack_rmse = mean_squared_error(y_test, stacking_pred)
stack_r2 = r2_score(y_test, stacking_pred)
print('Root Mean Squared Error: {0}'.format(stack_rmse))
print('R2: {0}'.format(stack_r2))   
```

In comparison to before, the RMSE for the STACKING MODEL is lower than the one for the Linear Regression, and the R2 is also slightly higher. This means, in the end, we will go for the STACKING MODEL.
